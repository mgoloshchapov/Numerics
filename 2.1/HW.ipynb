### Задача 5

```python
import seaborn as sns
from scipy.stats import pearsonr
```

```python
data = sns.load_dataset("anscombe")
gb = data.groupby(by="dataset")
datasets = [gb.get_group(x) for x in gb.groups]
```

```python
for df in datasets:
    x, y = df["x"], df["y"]
    plt.scatter(x, y)
    print("_____________________")
    print("sigma_x = {},\nsigma_y = {}, \npearson = {}".format(np.std(x), np.std(y), pearsonr(x, y)[0]))
    print("_____________________")
```

```python
for df, i in zip(datasets, range(1, 5)):
    plt.subplot(2,2,i)
    x, y = df["x"], df["y"]
    plt.scatter(x, y)
```

```python
import scipy.stats as sps


for df, i in zip(datasets, range(1, 5)):
    plt.subplot(2,2,i)
    x, y = df["x"], df["y"]
    k, b = np.polyfit(x, y, deg=1)
    plt.hist(y-k*x-b, bins=5)
    print(sps.shapiro(y-k*x-b))
```

### Задача 7

```python
plt.rcParams.update({"font.size" : 15, "font.family" : 'serif'})

x = np.linspace(0, 1, 10)

k, b = 1.0, 2.0
y0 = k*x + b
y = y0 + np.random.normal(loc=0.0, scale=0.1, size=len(x))

(k, b), cov = np.polyfit(x, y, deg=1, cov=True)


plt.scatter(x, y, label="data", color="red")
plt.plot(x, k*x + b, label ="fit, y = {}ч +", color="blue")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Fit example")

print(k, b, cov)
```
