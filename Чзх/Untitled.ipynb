```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from scipy.optimize import curve_fit
import pandas as pd
```

### Задача 5

```python
data = sns.load_dataset("anscombe")
data
```

```python
x1, y1 = data[data["dataset"]=="I"]["x"], data[data["dataset"]=="I"]["y"]
x2, y2 = data[data["dataset"]=="II"]["x"], data[data["dataset"]=="II"]["y"]
x3, y3 = data[data["dataset"]=="III"]["x"], data[data["dataset"]=="III"]["y"]
x4, y4 = data[data["dataset"]=="IV"]["x"], data[data["dataset"]=="IV"]["y"]
```

```python
plt.scatter(x1, y1)
plt.scatter(x2, y2)
plt.scatter(x3, y3)
plt.scatter(x4, y4)
```

```python
print("I: var(x)={}, var(y)={}".format(np.var(x1), np.var(y1)))
print("II: var(x)={}, var(y)={}".format(np.var(x2), np.var(y2)))
print("III: var(x)={}, var(y)={}".format(np.var(x3), np.var(y3)))
print("IV: var(x)={}, var(y)={}".format(np.var(x4), np.var(y4)))
```

```python
print(np.corrcoef(x1, y1))
print(np.corrcoef(x2, y2))
print(np.corrcoef(x3, y3))
print(np.corrcoef(x4, y4))
```

### Задача 7

```python
from sklearn import linear_model


x = np.linspace(0, 1, 10)
y = 5 * x + 2 + np.random.normal(loc=0.0, scale=0.5, size=10)

plt.scatter(x, y, color="black")
reg = linear_model.LinearRegression(fit_intercept=True)
reg.fit(x.reshape(-1, 1), y)


y_new = reg.predict(x.reshape(-1, 1))

plt.plot(x, y_new, color="red")
```
