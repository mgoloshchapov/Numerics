```python
# 8

def matmul(A, B):
    n = A.shape[0]
    C = [[0 for _ in range(n)] for _ in range(n)]
    for i in range(n):
        for j in range(n):
            for k in range(n):
                C[i][j] += A[i][k] * B[k][j]
    return C

Tmy = []
T = []
N = np.arange(1, 20)

time = %timeit -o -n 5 -r 5 -q 2 * 1
time.average

for n in N:
    a = np.random.randint(low=0, high=4, size=(n, n))
    b = np.random.randint(low=0, high=4, size=(n, n))
    
    time = %timeit -o -n 5 -r 5 -q matmul(a, b)
    Tmy.append(time.average)
    
    time = %timeit -o -n 50 -r 50 -q a@b
    T.append(time.average)
    
plt.figure(figsize=(10, 6))
plt.plot(N, Tmy, color='blue', label="matmul(a, b)")
plt.plot(N, T, color='red', label="a@b")
plt.xlabel("N")
plt.ylabel("Time, sec")
plt.tight_layout()

plt.legend()

plt.figure(figsize=(10, 6))
plt.subplot(121)
plt.plot(N, Tmy, color='blue', label="matmul(a, b)")
plt.xlabel("N")
plt.ylabel("Time, sec")
plt.subplot(122)
plt.plot(N, T, color='red', label="a@b")
plt.tight_layout()
plt.xlabel("N")
plt.ylabel("Time, sec")
plt.legend()
```

### Упр 3 Вигнеровский полукруг

```python
n = 400

H = []
bins = []

for _ in range(50):
    A = np.random.normal(loc=0.0, scale=1.0, size=(n, n))
    B = A + A.T
    hist, bin_edges = np.histogram(np.linalg.eigvals(B), bins=30)
    H.append(hist)
    bins.append(bin_edges)

bins = np.array(bins).mean(axis=0)
H = np.array(H)
bar = H.mean(axis=0)
err = H.std(axis=0)

print(bins.shape)
print(bins)
print(bar.shape)
plt.fill_between(bins[:-1],bar-err, bar+err, alpha=0.3,color="red")
plt.plot(bins[:-1], bar, color="red", linewidth=3)
```
