```python
%matplotlib ipympl
import matplotlib.pyplot as plt
import numpy as np
import ipympl
from scipy.sparse.linalg import svds

plt.rcParams.update({"font.size" : 15, "font.family" : "serif"})
```

### Задача 1

```python
A = np.random.random(size=(3, 3))
A = A.T@A


#Очень круто, что плоскость проходит через ноль, так как mean=0
N = 100
X = np.random.multivariate_normal(mean=np.zeros(3), cov=A, size=N).T

u, s, vh = svds(X, 2)
X1 = u@np.diag(s)@vh
x, y, z = X
```

```python
u
```

```python
from mpl_toolkits.mplot3d import Axes3D


fig = plt.figure()
ax = Axes3D(fig)
  
plot = ax.scatter(x, y, z, color='blue')
ax.quiver([0, 0], [0, 0], [0, 0], u[0]@np.diag(s**0.5), u[1]@np.diag(s**0.5), u[2]@np.diag(s**0.5))
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")


v1, v2 = u.T
t1, t2 = np.arange(-3, 3, 0.1), np.arange(-3, 3, 0.1)
r = np.array([x, y, z]).T


r_new = [r[-1]]
for tt1 in t1:
    for tt2 in t2:   
        r_new.append(r[1] + tt1 * v1 + tt2 * v2)
  

xx, yy, zz = np.array(r_new).T
ax.scatter(xx, yy, zz, color="blue", s=5, alpha=0.05)
ax.scatter(x, y, z, color="red", s=5)
```

### Задача 6

```python
# Генерим точки
from scipy.stats import norm

mu1 = 0 
mu2 = 10
s1 = 1
s2 = 2
a = 0.5
n = 1000


def p(x, mu1, s1, mu2, s2, a):
    return a*norm.pdf(x, loc=mu1, scale=s1)+(1-a)*norm.pdf(x, loc=mu2, scale=s2)

x = np.linspace(-10, 20, 200)
sample = np.random.choice(x, size=n,p=p(x, mu1, s1, mu2, s2, a)/np.sum(p(x, mu1, s1, mu2, s2, a)))



fig, ax = plt.subplots()
ax.plot(x, p(x, mu1, s1, mu2, s2, a))
ax.set_xlabel("X")
ax.set_title("PDF")
```

```python
fig, ax = plt.subplots()
ax.hist(sample, bins=40)
ax.set_xlabel("X")
ax.set_ylabel("N samples")
```
