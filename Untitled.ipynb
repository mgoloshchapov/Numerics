```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import plotly.express as px

plt.rcParams.update({"font.size" : 15, "font.family" : "serif"})
```

```python
data1 = pd.read_csv("Trace_0002.csv", skiprows=45)
data2 = pd.read_csv("Trace_0000.csv", skiprows=45)
data2
```

```python
x1, y1 = data1.iloc[::, 0], data1.iloc[::, 1]
x2, y2 = data2.iloc[::, 0], data2.iloc[::, 1]
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(121)
a, b = 200,100 
plt.title("Unlocked")
plt.plot(x1[a:-b], y1[a:-b], color="black")
plt.xlabel("Hz")
plt.ylabel("dBm")

plt.subplot(122)
plt.title("Locked")
plt.plot(x2, y2, color="black")
plt.xlabel("Hz")
plt.ylabel("dBm")

plt.tight_layout()
```

```python
Ref = 3.417 * 10**9

plt.figure(figsize=(12, 6))
plt.subplot(121)
a, b = 460,500
plt.title("Unlocked")
plt.plot((x1-Ref)[a:-b], 10**(y1)[a:-b], color="black")
plt.xlabel("Beat-note freq - Ref, Hz")
plt.ylabel("V")
plt.yticks([])

plt.subplot(122)
a, b = 480,480
plt.title("Locked")
plt.plot((x2-Ref)[a:-b],10**(y2)[a:-b], color="black")
plt.xlabel("Hz")
plt.ylabel("dBm")

plt.tight_layout()
```
